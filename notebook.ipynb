```python
def tts_transformer(script):
    script=script.replace('\n',' ')
    ls = []
    while len(script) >= 75:
        chunk = script[:75]
        for sep in ['।', '?', '…', '—', ',', ' ']:
            index=chunk.rfind(sep)
            if index>0:
                ls.append(chunk[:index+1].strip())
                script=script[index+1:]
                break
        else:
            ls.append(chunk.strip())
            script=script[76:]
    return ls
```

```python
script='''
एक युवक सुबह अचानक एक अजनबी घर में जागता है। आईने में झाँकता है, तो देखता है कि वह कोई और है — उम्र, चेहरा, सब कुछ बदला हुआ।
वह घबरा जाता है। मोबाइल उठाकर नाम देखने की कोशिश करता है, लेकिन फोन लॉक है। कमरे की दीवारों पर अनजान लोगों की तस्वीरें टंगी हैं।
घर का दरवाज़ा खोलते ही उसे एक महिला ‘बेटा’ कहकर आवाज़ देती है। युवक चुप रह जाता है। उसे कुछ समझ नहीं आता कि ये लोग कौन हैं।
वह जल्दी-जल्दी नाश्ता करता है और बाहर निकल जाता है, सोचते हुए — "ये सब क्या हो रहा है?" बाहर की दुनिया भी उसके लिए नई है।
शाम होते-होते, वह एक पार्क में बैठा सोच रहा होता है, तभी एक अजनबी बच्चा आकर कहता है — “कल तुम और जैसे नहीं दिखोगे।” युवक चौंक जाता है।
अगली सुबह… उसकी आँखें खुलती हैं — अब वह एक लड़की के शरीर में है। आईने में देखता है तो उसकी साँस अटक जाती है।
अब सवाल ये है — कौन है ये? क्यों हर दिन नया शरीर? और उस बच्चे ने कैसे जान लिया? जानने के लिए अगले भाग का इंतज़ार कीजिए। 
हमारे चैनल को लाइक, शेयर और सब्सक्राइब ज़रूर करें… अगला भाग जल्दी आ रहा है।

अब वह एक लड़की के शरीर में है। चलने-फिरने से लेकर बोलने तक सब अलग लगता है। वह आईने में देखती है और फुसफुसाती है — "मैं कौन हूँ?"
अचानक उसकी माँ कमरे में आती है — “बिटिया, देर हो रही है स्कूल के लिए।” वो झूठमूठ मुस्कुरा देती है, पर अंदर ही अंदर डर से काँप रही होती है।
स्कूल की गलियों में चलते हुए उसे सब कुछ नया लगता है। दोस्त उसे बुलाते हैं, लेकिन वह किसी को नहीं पहचानती।
क्लास में बैठकर वह चुपचाप दीवार घड़ी को घूरती रहती है। वही टूटी हुई घड़ी — जिसमें अब भी वही वक्त अटका है: चार बजकर बारह मिनट।
टीचर कुछ पूछते हैं, पर उसे जवाब नहीं आता। वो बड़बड़ाती है — “कहीं ये सपना तो नहीं?” तभी बगल में बैठा छात्र कहता है — “तू भी देख रही है न वो घड़ी?”
छात्र फुसफुसाता है — “ये घड़ी हर उस इंसान को दिखती है जो फँसा हुआ है…” और तभी घंटी बजती है।
अब सवाल और गहराते हैं — क्या ये लड़का भी जानता है कुछ? क्या और लोग भी फँसे हैं इस रहस्य में?
अगला भाग ज़रूर देखिए। चैनल को लाइक, शेयर और सब्सक्राइब करें ताकि आप कुछ भी मिस न करें।

अगली सुबह जब उसकी आँख खुलती है, तो वह अब एक बूढ़े आदमी के शरीर में होता है। कमरे में एक छड़ी, दवाइयाँ और पुराने ज़माने की चीज़ें पड़ी हैं।
आईने में अपनी झुर्रियों भरी शक्ल देखकर वह फुसफुसाता है — “अब क्या मैं बूढ़ा हूँ?” पर मन अब भी वही पुराना है — उलझन और बेचैनी से भरा।
वह पुराने अलमारी की तलाशी लेता है और एक छोटा लकड़ी का डिब्बा निकालता है — जिसके अंदर है वही टूटी घड़ी… और एक पुराना फ़ोटो।
फ़ोटो में वही बच्चा है जिसने पार्क में उसे चेतावनी दी थी। लेकिन यह फ़ोटो तो बहुत पुरानी लगती है — कम से कम बीस साल पुरानी।
घड़ी अचानक खुद-ब-खुद चलने लगती है, और कमरे की दीवारों पर अजीब आकृतियाँ बनने लगती हैं — जैसे कोई सपना हकीकत में बदल रहा हो।
वह घबरा जाता है, पर उसी क्षण, दीवार पर एक दृश्य उभरता है — हॉस्पिटल का एक कमरा, जिसमें कोई कोमा में लेटा है। चेहरा दिखता नहीं… पर एहसास होता है, 'वो मैं हूँ।
अब वो यह सोचने पर मजबूर है — क्या ये सब सपना है? क्या वह असल में कोमा में है? या कोई और खेल चल रहा है?
इस रहस्य से पर्दा हटेगा अगले भाग में। चैनल को लाइक करें, शेयर करें और सब्सक्राइब करना न भूलें।

इस बार वह एक सरकारी अफ़सर के शरीर में जागता है। सामने मेज़ पर फाइलें हैं, और दीवार पर देश का झंडा टंगा है। सब कुछ बड़ा, पर भारी-सा लगता है।
एक फाइल में वही घड़ी की तस्वीर छपी होती है। शीर्षक होता है — ‘प्रोजेक्ट समय-फाँस’। वह पढ़ते-पढ़ते कांप उठता है।
फाइल में लिखा होता है — ‘कुछ मस्तिष्क ऐसे हैं जो सपनों में फँस जाते हैं… और वहीं नई दुनिया रच लेते हैं।’
वह भागता है — गलियारों से, सीढ़ियों से… और एक गुप्त तहख़ाने में पहुँचता है। वहाँ एक पुराना कम्प्यूटर चलता हुआ मिलता है।
स्क्रीन पर वही बच्चा फिर दिखाई देता है — इस बार सीधा कैमरे की ओर देखकर कहता है — “तुम अब भी नहीं जागे… तुम्हें सच्चाई से डर लगता है।”
कमरे की दीवारें फटने लगती हैं, घड़ी की टिक-टिक तेज़ हो जाती है, और आसपास की दुनिया पिघलने लगती है — जैसे सपने का पर्दा हट रहा हो।
अब सच बस एक क़दम दूर है। पर क्या वह सामना कर पाएगा? या वह हमेशा के लिए इसी झूठ के शहर में खो जाएगा?
इसका उत्तर मिलेगा अंतिम भाग में। चैनल को लाइक करें, शेयर करें और सब्सक्राइब करना न भूलें।

वह अब एक सुनसान अस्पताल के कमरे में जागता है। खिड़कियों पर पर्दे हैं, हर तरफ सन्नाटा है, और मशीनों की धीमी बीप सुनाई दे रही है।
आईने की जगह दीवार पर वही टूटी घड़ी टंगी है — अब उसमें सिर्फ एक सेकंड बचा है।
घड़ी की टिक-टिक थम जाती है। तभी पूरे कमरे में सफेद रोशनी फैल जाती है, और उसे अपने बचपन की आवाज़ें सुनाई देने लगती हैं — “उठो बेटा…”
अचानक उसकी आँखों के सामने एक खिड़की खुलती है — वह खुद को एक अस्पताल के बेड पर लेटा देखता है। मशीनें चल रही हैं। असली शरीर वहीं है।
अब वह समझ जाता है — ये सब सपना था, एक लंबा, उलझा हुआ, पर गहरा सपना। वह आँखें बंद करता है… और फिर धीरे-धीरे खोलता है।
वास्तविकता में लौटकर उसकी आँखों से आँसू बह जाते हैं। वह मुस्कराता है… ज़िंदा है। सपना टूट चुका है।
कभी-कभी सबसे बड़ा रहस्य यही होता है — *क्या हम जागे हैं या अब भी सपने में?*
अगर आपको ये कथा पसंद आई, तो कृपया लाइक करें, शेयर करें, और सब्सक्राइब करना न भूलें। अगली कहानी में फिर मिलेंगे।


एक घना जंगल था, जहाँ एक साफ़-सुथरी झील भी थी। उस झील में एक कछुआ रहता था जिसका नाम धीरू था। धीरू थोड़ा धीरे चलता था, पर बहुत समझदार था। जंगल में दो बगुलों से धीरू की गहरी दोस्ती थी, उनके नाम थे चंचल और पंखुड़ी।
एक बार जंगल में भयंकर सूखा पड़ गया। झील का पानी सूखने लगा। धीरू और उसके दोस्त बहुत परेशान थे।
चंचल बोला, "हमें कहीं और जाना होगा, जहाँ पानी हो!" पंखुड़ी ने कहा, "पर तुम कैसे जाओगे, धीरू? हम तो उड़ सकते हैं!"
धीरू ने थोड़ा सोचा और एक उपाय बताया। "तुम दोनों एक मज़बूत लकड़ी अपनी-अपनी चोंच में पकड़ लेना, और मैं बीच में से उसे अपने मुँह से पकड़ लूँगा।"
बगुले एक-दूसरे को देखने लगे। पंखुड़ी ने चेतावनी दी, "लेकिन धीरू, अगर तुमने मुँह खोला, तो नीचे गिर जाओगे!" धीरू ने वादा किया कि वह बिल्कुल नहीं बोलेगा।
बगुले मान गए। उन्होंने लकड़ी उठाई, और धीरू ने भी उसे अपने मुँह से मज़बूती से पकड़ लिया। फिर वे धीरे-धीरे उड़ना शुरू हुए।
वे जंगल के ऊपर से उड़ रहे थे। नीचे गाँव के लोग, बच्चे और जानवर उन्हें देखकर हैरान थे। "देखो, देखो! कछुआ उड़ रहा है!" लोग चिल्लाने लगे।
यह सुनकर धीरू को बहुत गुस्सा आया। वह लोगों को जवाब देना चाहता था। जैसे ही उसने कुछ बोलने के लिए अपना मुँह खोला, वह लकड़ी से छूटा और धड़ाम से नीचे जा गिरा!
खुशकिस्मती से, वह घास के ढेर पर गिरा, इसलिए उसे ज़्यादा चोट नहीं आई। बगुले भी तुरंत उसके पास उतर आए।
"धीरू," चंचल उदास होकर बोला, "हमने तुम्हें मना किया था बोलने से।" पंखुड़ी ने कहा, "तुम्हारी एक छोटी सी गलती से सब बेकार हो गया।"
धीरू को अपनी गलती का एहसास हुआ। उसने उदास होकर कहा, "तुम ठीक कह रहे हो दोस्तों। मुझे अपने गुस्से पर काबू रखना चाहिए था। मैंने अपनी बुद्धिमानी का इस्तेमाल नहीं किया।"


एक अनोखा सफर
एक छोटे से गाँव में रहता था रमेश, एक साधारण सा लड़का। उसका सपना था शहर जाकर कुछ बड़ा करना, लेकिन उसकी जेब में सिर्फ़ पचास रुपये थे। एक दिन, उसने हिम्मत जुटाई और बस में बैठकर शहर की ओर चल पड़ा। बस में बैठा एक बूढ़ा व्यक्ति उससे बात करने लगा। उसने रमेश से पूछा, "बेटा, कहाँ जा रहे हो?"
रमेश ने शर्माते हुए कहा, "शहर, कुछ काम ढूँढने।" बूढ़े व्यक्ति ने मुस्कुराकर कहा, "सपने बड़े होने चाहिए, लेकिन रास्ता आसान नहीं होता।" फिर उसने अपनी जेब से एक पुराना लाकेट निकाला और रमेश को दे दिया। "ये रखो, ये तुम्हें हिम्मत देगा," उसने कहा।
शहर पहुँचकर रमेश को कई मुश्किलें आईं। नौकरी नहीं मिली, रहने की जगह नहीं थी। एक रात, ठंड में सड़क पर बैठा वह लाकेट को देखने लगा। उसमें एक छोटा सा नक्शा था, जो एक पुराने मंदिर की ओर इशारा करता था। उत्सुकता में रमेश उस मंदिर की तलाश में निकल पड़ा।
मंदिर में उसे एक बूढ़ी माँ मिलीं, जो वहाँ पूजा करती थीं। रमेश ने अपनी कहानी सुनाई। माँ ने कहा, "बेटा, असली खजाना मेहनत और विश्वास में है।" उन्होंने रमेश को अपने छोटे से ढाबे में काम करने का मौका दिया। रमेश ने दिन-रात मेहनत की और जल्द ही ढाबा मशहूर हो गया। उसकी मेहनत रंग लाई, और उसने अपना खुद का ढाबा खोल लिया।
एक दिन, वही बूढ़ा व्यक्ति ढाबे पर आया। रमेश ने उसे पहचान लिया और लाकेट लौटाने की कोशिश की। बूढ़े ने हँसकर कहा, "ये तो बस एक बहाना था, असली जादू तो तुममें था।"
संदेश: सपने पूरे करने के लिए मेहनत और विश्वास ही असली खजाना है।


खोया हुआ रहस्य
गंगा के किनारे बसे एक छोटे गाँव में रहती थी अनन्या, एक जिज्ञासु लड़की। उसे पुरानी कहानियाँ और रहस्य बहुत पसंद थे। एक दिन, गाँव के पुराने हवेली के पास खेलते हुए उसे एक चमकदार, तांबे का सिक्का मिला, जिस पर अजीब-से निशान थे। सिक्के के पीछे लिखा था, "रास्ता वही जो हिम्मत दिखाए।"
अनन्या ने गाँव के बुजुर्ग पंडित जी को सिक्का दिखाया। उन्होंने आँखें चौड़ी करते हुए कहा, "ये सिक्का गाँव के खोए हुए खजाने का हिस्सा है! ये हवेली में छिपा है, लेकिन रास्ता खतरनाक है।" अनन्या का दिल जोश से भर गया। उसने ठान लिया कि वह खजाना ढूँढेगी।
रात को, जब सब सो रहे थे, अनन्या चुपके से हवेली में घुसी। वहाँ अंधेरा और सन्नाटा था। सिक्के पर बने निशानों को देखकर उसने दीवार पर एक छिपा हुआ बटन दबाया। अचानक, फर्श खिसका और एक गुप्त सीढ़ी दिखाई दी। डरते-डरते वह नीचे उतरी। वहाँ उसे एक पुराना संदूक मिला, लेकिन ताला बंद था। ताले पर वही निशान थे जो सिक्के पर थे।
अनन्या ने सिक्का ताले में डाला, और संदूक खुल गया। अंदर थीं पुरानी किताबें, गहने और एक पत्र। पत्र में लिखा था, "ये खजाना गाँव की भलाई के लिए है।" अनन्या ने खजाना गाँव वालों को सौंप दिया। गाँव में स्कूल, अस्पताल और सड़कें बनीं। अनन्या की हिम्मत की हर कोई तारीफ करने लगा।
एक दिन, पंडित जी ने अनन्या से कहा, "सिक्का तो सिर्फ़ एक संकेत था। असली खजाना तुम्हारी हिम्मत और नेकदिली थी।"
संदेश: हिम्मत और अच्छे इरादे किसी भी रहस्य को सुलझा सकते हैं।


एक दिन का मेहमान
एक छोटे से शहर में रहती थी काव्या, एक चुलबुली और सपने देखने वाली लड़की। उसका जीवन सामान्य था, लेकिन एक सुबह कुछ अजीब हुआ। जब वह जागी, तो उसे एहसास हुआ कि वह अपने शरीर में नहीं, बल्कि अपनी सहेली रिया के शरीर में थी! काव्या हैरान थी, लेकिन उसने सोचा शायद यह एक सपना है। उसने रिया की तरह दिन बिताया, स्कूल गई, दोस्तों से मिली, और रिया के परिवार के साथ समय बिताया। रिया की जिंदगी इतनी अलग थी—उसके माता-पिता सख्त थे, और उसे पढ़ाई का बहुत दबाव था।
अगले दिन, काव्या फिर अपने शरीर में नहीं थी। इस बार वह अपने स्कूल के शांत लड़के, अर्जुन, के शरीर में थी। अर्जुन की दुनिया बिल्कुल अलग थी—वह चुपके से कविताएँ लिखता था और अपने बीमार दादाजी की देखभाल करता था। काव्या को उसकी संवेदनशीलता ने छू लिया। हर दिन, वह किसी नए व्यक्ति के शरीर में जागती—कभी अपनी टीचर, कभी दुकानदार, तो कभी एक अनजान लड़के के।
पहले तो काव्या डर गई, लेकिन धीरे-धीरे उसे समझ आया कि यह अनुभव उसे दूसरों की जिंदगी को समझने का मौका दे रहा था। उसने हर दिन को खास बनाने की ठानी। रिया के शरीर में, उसने रिया के माता-पिता से खुलकर बात की और उन्हें उसकी इच्छाएँ बताईं। अर्जुन के शरीर में, उसने उसकी एक कविता स्कूल के बोर्ड पर चिपकाई, जिसे सबने सराहा। हर व्यक्ति के जीवन में वह कुछ अच्छा छोड़ने लगी।
एक दिन, काव्या एक रहस्यमयी पत्र में जागी, जिसमें लिखा था, "जिंदगी को दूसरों के नजरिए से देखो, तभी सच्चा प्यार और खुशी मिलेगी।" उस दिन वह फिर अपने शरीर में थी। काव्या ने फैसला किया कि वह अब हर किसी से प्यार और समझदारी से मिलेगी, क्योंकि उसने देखा था कि हर इंसान की अपनी कहानी होती है।


"कभी सोचा है अगर आपकी परछाईं ज़िंदा हो जाए… और वो करे वही, जो आप चाहते हो? नहीं? तो सुनो ये कहानी… अर्जुन की।"
अर्जुन, एक सीधा-सादा लड़का था, दिल्ली की भीड़ में खोया हुआ। ऑफिस से घर, घर से ऑफिस - यही उसकी दुनिया थी। लेकिन एक दिन, कुछ ऐसा हुआ जिसने उसकी ज़िंदगी बदल दी।
रात के अंधेरे में, जब अर्जुन अपने कमरे में अकेला बैठा था, बिजली चली गई। जैसे ही उसने टॉर्च ऑन की, उसकी परछाईं दीवार पर कुछ अजीब सी हरकतें करने लगी। वो हिली… मुड़ी… और फिर अर्जुन से बोल पड़ी:
“मैं तेरी छाया हूं… और अब तेरी शक्ति।”
अर्जुन डर गया, लेकिन छाया ने कहा, “जो तू सोचता है, मैं कर सकती हूं। चोरी, बचाव, उड़ान… जो तू चाहे।”
अर्जुन ने शक के तौर पर कहा, "अगर तू सच में शक्ति है… तो ऑफिस के उस बॉस की फाइल चुपचाप ले आ जो मेरी तरक्की रोक रहा है।"
अगली सुबह, वो फाइल अर्जुन की मेज़ पर थी।
अब अर्जुन ने इस्तेमाल शुरू किया - ट्रैफिक से बचना? छाया ने सब सिग्नल ग्रीन कर दिए। ATM से पैसे? छाया ने बिना कार्ड के दे दिए। वो सब करने लगा जो वो सिर्फ सपनों में सोचता था।
लेकिन फिर आया असली मोड़…
एक रात छाया ने खुद कहा - “अब मेरी भी इच्छाएँ हैं अर्जुन… अब मैं तुझसे आज़ाद होना चाहती हूं।”
अर्जुन हैरान रह गया। “तू मेरी शक्ति है! मेरी मदद करती है!”
छाया बोली - “मैं मदद नहीं कर रही… मैं तुझे बदल रही हूं। इंसान से लालची राक्षस बना रही हूं।”
और फिर छाया ने खुद को अर्जुन से अलग कर लिया। अब वो एक आज़ाद अस्तित्व थी… जिसकी खुद की मर्ज़ी थी। अर्जुन फिर से अकेला था… पर अब वो जान चुका था कि हर शक्ति एक कीमत के साथ आती है।
"अगर तुम्हारे पास ये शक्ति होती… कि तुम्हारी छाया तुम्हारे हर इशारे पर चले… तो क्या तुम उससे दोस्ती करते या गुलामी?"
"कमेंट में ज़रूर बताना… और ऐसे ही कहानियों के लिए सब्सक्राइब करना मत भूलना!"
अगर आप चाहें, मैं इसका स्क्रिप्ट फ़ॉर्मेट, थंबनेल आइडिया, और बैकग्राउंड म्यूजिक के सुझाव भी दे सकता हूँ। बताइए?


"लॉटरी का शाप"
"पैसे सबको चाहिए… लेकिन क्या हर पैसा किस्मत से आता है? या कभी-कभी… बदकिस्मती बनकर?"
ये कहानी है राघव की — एक मामूली मिडिल क्लास आदमी, जो हर महीने अपनी सैलरी के बाद एक ही सपना देखता था… "अगर एक दिन लॉटरी लग गई, तो…"
एक दिन, बस स्टैंड पर राघव को एक फटे-पुराने कपड़े वाला बूढ़ा आदमी मिला। उसने कहा:
“ये टिकट लो… ये खास है। किस्मत खुद चलकर आएगी।"
राघव हँसा, लेकिन टिकट सिर्फ ₹ दस का था। ले लिया।
अगले दिन… न्यूज में आया - "विजेता टिकट नंबर: सात आठ छह पाँच दो नौ"
राघव की आंखें फटी की फटी रह गईं… वो वही टिकट था। इनाम - पच्चीस करोड़ रुपये।
खुशियाँ! सेलिब्रेशन! नया बंगला, नई कार, पुरानी जॉब को बाय-बाय!
लेकिन… फिर कुछ अजीब होने लगा।
पहले दिन - उसकी बीवी को एक खौफनाक सपना आया, कि किसी ने उनका पीछा किया।
दूसरे दिन - उसका दोस्त जो उसे बधाई देने आया था, एक्सीडेंट में मर गया।
तीसरे दिन - राघव की कार ब्रेक फेल हो गई… और वो बाल-बाल बचा।
अब राघव को उस बूढ़े की याद आई। वो फिर उसी बस स्टैंड गया। बूढ़ा वहीं था… जैसे उसका इंतज़ार कर रहा हो।
"तू सोचता है मैंने तुझे इनाम दिया?"
"नहीं… ये एक परीक्षा थी। तू भी दूसरों की तरह लालच में फँस गया।”
राघव घबरा गया - "मैं क्या करूं?"
बूढ़ा बोला -
"इस लॉटरी की कीमत है… जो तू चुकाता रहेगा… हर उस चीज़ से जो तुझे प्यारी है।”
और फिर… वो बूढ़ा गायब हो गया।
राघव ने अगले दिन वो टिकट फाड़ दिया… बैंक अकाउंट खाली कर दिया… और सब कुछ छोड़कर हिमालय चला गया। कहते हैं… आज भी वो शांति की तलाश में भटक रहा है।
"अगर आपको कल कोई कहे कि ये लॉटरी आपको करोड़पति बना देगी… क्या आप वो टिकट लेंगे?"
"सोचिए… और कॉमेंट में बताइए। अगली कहानी में मिलते हैं!"
'''
```
